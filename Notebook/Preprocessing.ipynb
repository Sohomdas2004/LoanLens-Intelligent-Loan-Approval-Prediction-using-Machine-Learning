```python
import pandas as pd
df=pd.read_csv('/content/loan.csv')
```

```python
df.head()
```

```python
df.drop('Loan_ID',axis=1)
```

```python
df.dropna(inplace=True)
```

```python
x=df.drop(['Loan_Status','Loan_ID'],axis=1)
y=df['Loan_Status']
```

```python
x['Dependents'] = x['Dependents'].replace('3+', 3)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
le=OrdinalEncoder()
y=le.fit_transform(y.values.reshape(-1,1))
```

```python
categorical_cols=['Gender','Married','Education','Self_Employed','Property_Area']
numerical_cols=['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
ct = ColumnTransformer(
    transformers=[
        ('onehot',OrdinalEncoder(), categorical_cols),
        ('scaler', StandardScaler(), numerical_cols)
    ]
)
```

```python
x_train_transformed = ct.fit_transform(x_train)
x_test_transformed = ct.transform(x_test)
```
